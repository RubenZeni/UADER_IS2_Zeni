```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt

# --- Definicin de las funciones de clculo ---
# Estas funciones se basan en las relaciones dadas en el enunciado [cite: 514]

def calcular_esfuerzo(S):
    """Calcula el esfuerzo (E) basado en el tamao del proyecto (S)."""
    # E = 8 * S^0.95 [cite: 515]
    return 8 * (S**0.95)

def calcular_tiempo(E):
    """Calcula el tiempo calendario (td) basado en el esfuerzo (E)."""
    # d = 2.4 * E^0.33 [cite: 516]
    return 2.4 * (E**0.33)

# --- Grfico 1: Esfuerzo (E) vs. Tamao (S) ---

# Generar un rango de valores para el tamao S de 0 a 10000 [cite: 513]
# Usamos np.linspace para crear 100 puntos en ese intervalo para una curva suave
S_vals = np.linspace(0, 10000, 100)

# Calcular los valores de esfuerzo correspondientes para cada tamao
E_vals = calcular_esfuerzo(S_vals)

# Crear el grfico
plt.figure(figsize=(10, 6)) # Define el tamao de la figura
plt.plot(S_vals, E_vals, label='E = 8 * S^0.95', color='blue')
plt.title('Esfuerzo del Proyecto vs. Tamao')
plt.xlabel('Tamao del Proyecto (S)')
plt.ylabel('Esfuerzo (E)')
plt.grid(True) # Agrega una grilla para mejor lectura
plt.legend()
plt.savefig('esfuerzo_vs_tamano.png') # Guarda el grfico como imagen
plt.show() # Muestra el grfico

# --- Grfico 2: Tiempo (td) vs. Esfuerzo (E) ---

# Generar un rango de valores para el esfuerzo E de 1 a 500 [cite: 513]
E_intervalo = np.linspace(1, 500, 100)

# Calcular los valores de tiempo correspondientes para cada esfuerzo
td_vals = calcular_tiempo(E_intervalo)

# Crear el segundo grfico
plt.figure(figsize=(10, 6))
plt.plot(E_intervalo, td_vals, label='d = 2.4 * E^0.33', color='green')
plt.title('Tiempo Calendario vs. Esfuerzo')
plt.xlabel('Esfuerzo (E)')
plt.ylabel('Tiempo Calendario (d)')
plt.grid(True)
plt.legend()
plt.savefig('tiempo_vs_esfuerzo.png') # Guarda el grfico como imagen
plt.show()
```
